## Plot - Basic

```python
# 그림(figure)과 축(axes) 생성
fig=plt.figure()
ax=plt.axes()
```

```python
x=np.arange(0,10,1)
x
```

```python
plt.plot(x,x+1,color='red')       # red
plt.plot(x,x+2,color='blue')      # Blue
plt.plot(x,x+3,color='0.5')       # between 0 and 1, grey color
plt.plot(x,x+4,color='#FFDD44')   # RGB code
```

```python
# 플롯 그리드 생성
fig,ax=plt.subplots()

# plot background
plt.style.use('seaborn-whitegrid')

# plot
ax.plot(x,x+1,linestyle='solid', label='y=x+1')    
ax.plot(x,x+2,linestyle='dashed', label='y=x+2')      
ax.plot(x,x+3,linestyle='dashdot', label='y=x+3')     
ax.plot(x,x+4,linestyle='dotted', label='y=x+4')   

# Label & Legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='lower right')
```
